```python
raw_text = """Since its founding in 1939 as Timely Comics, the Marvel Universe has grown into a sprawling fictional cosmos where superheroes, villains, cosmic entities, and ordinary people intersect in stories that blend action, humor, and human drama. At its core, Marvel emphasizes character-driven narratives: heroes like Spider-Man struggle with personal responsibility; the X-Men confront prejudice and identity; Iron Man grapples with technology and morality. This focus on flawed, relatable protagonists helped distinguish Marvel comics from earlier, more archetypal fare and enabled stories that explore social issues, ethical dilemmas, and the costs of heroism.
The Marvel Universe is also notable for its interconnectedness. Characters cross paths frequently, events in one title ripple into others, and large crossover sagas reshape continuity, creating a sense of a living, evolving world. Major storylines — from civil conflicts to multiversal crises — provide spectacle while allowing creators to examine political, philosophical, and emotional themes. The setting spans earthly neighborhoods like Manhattan’s Hell’s Kitchen and Wakanda’s hidden vibrancy to cosmic vistas where the Celestials and the Infinity Stones govern fate, offering a canvas broad enough for intimate street-level tales and operatic cosmic epics alike.
Another defining trait is Marvel’s commitment to reinvention. Over decades, writers and artists have reinterpreted characters, introduced legacy heroes, and experimented with genres, keeping the universe relevant across generations. The editorial approach balances respect for continuity with bold new directions: legacy characters inherit mantles, alternative timelines permit risk-taking, and periodic reboots and events refresh the mythos. This adaptability is why decades-old characters continue to feel contemporary, resonating with new audiences while retaining core identities.
Beyond comics, the Marvel Universe has achieved unprecedented global reach through film, television, animation, and merchandise. The Marvel Cinematic Universe (MCU) translated serialized comic instincts into interlocking blockbuster films and shows, popularizing shared-universe storytelling in mainstream media. This multimedia expansion amplified Marvel’s cultural impact, turning iconic lines, costumes, and phrases into shared cultural references and inspiring diverse fandom communities. Simultaneously, independent comics, animated series, and critically acclaimed graphic novels demonstrate the brand’s artistic range.
At its best, the Marvel Universe combines high-concept imagination with grounded human stories. It asks what it means to live ethically in a world of extraordinary power, and how communities, institutions, and individuals respond to threat, belonging, and change. Whether through the street-level grit of Daredevil, the cosmic odysseys of the Guardians of the Galaxy, or political intrigue surrounding S.H.I.E.L.D., Marvel remains a vibrant storytelling laboratory. Its enduring appeal lies in its capacity to evolve while still honoring the emotional core of its characters — the personal struggles, the friendships, the losses, and the hope that drive heroes to act.
Through decades of creativity and commercial growth, Marvel has also fostered critical conversations about representation, authorship, and adaptation. Modern creators continue expanding the universe with diverse voices and formats, exploring marginalized perspectives and innovative narrative techniques. As technologies and audiences change, Marvel’s stories adapt, ensuring the universe remains both a reflection of its era and an imaginative space for future storytellers and fans and creators everywhere."""


all_words = sorted(set(raw_text.split()))    

vocab_size = len(all_words)

all_words
```

```python
vocab = { token: integer for integer, token in enumerate(all_words) }
print(f"Vocab size: {vocab}")

for i, item in enumerate(vocab.items()):
    print(f"{i}: {item}")
    if i >=50:
        break
```

```python
import re
class SimpleTokenizerV1:
    def __init__(self, vocab):
       self.str_to_int = vocab
       self.int_to_str = { integer: token for token, integer in vocab.items() }
    
    def encode(self, text):
        preprocessed = re.split(r'([,.?_!"()\']|--|\s+)', text)
        preprocessed = [token for token in preprocessed if token.strip() != '']
        return [self.str_to_int[token] for token in preprocessed if token in self.str_to_int]
    
    def decode(self, token_ids):
        text = " ".join([self.int_to_str[token_id] for token_id in token_ids])
        text = re.sub(r'([,.?_!"()\']|--)', r' \1 ', text)

        return re.sub(r'\s+', ' ', text).strip()
        

token = SimpleTokenizerV1(vocab)
```

```python
encode_text = token.encode(raw_text)
encode_text[:50]
```

```python
token.decode(encode_text)
```

```python
class SimpleTokenizerV2:
    def __init__(self, vocab):
        self.str_to_int = dict(vocab)
        if "<|unk|>" not in self.str_to_int:
            self.str_to_int["<|unk|>"] = len(self.str_to_int)
        self.int_to_str = {i: s for s, i in self.str_to_int.items()}

    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [
            item.strip() for item in preprocessed if item.strip()
        ]
        preprocessed = [item if item in self.str_to_int            #1
                        else "<|unk|>" for item in preprocessed]

        ids = [self.str_to_int[s] for s in preprocessed]
        return ids

    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)    #2
        return text


tokenizerv2 = SimpleTokenizerV2(vocab)
encodeV2 = tokenizerv2.encode(raw_text)
encodeV2[:50]
```

```python
tokenizerv2.decode(encodeV2)
```
